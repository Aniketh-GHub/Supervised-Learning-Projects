```python
import numpy as np
from sklearn.model_selection import train_test_split

# Assuming 'data' is your dataframe and 'Age' is the target variable
y = data["Assets"].values  # Set your target variable (e.g., "Assets")

# Define your feature set (excluding the target 'Assets' column)
x_data = data.drop(["Assets"], axis=1)

# Selecting only numeric columns from x_data
x_data = x_data.select_dtypes(include=[np.number])

# Normalize the data
x = (x_data - np.min(x_data)) / (np.max(x_data) - np.min(x_data))

# Splitting the dataset into training and testing sets
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=1)

# Check the shapes
print(x_train.shape, x_test.shape)
```

```python
from sklearn.svm import SVC
svm=SVC(random_state=1)
svm.fit(x_train,y_train)
print("train accuracy:",svm.score(x_train,y_train))
print("test accuracy:",svm.score(x_test,y_test))
```
