```python
X = df.drop('Exited', axis=1) #independent
y = df['Exited'] #dependent
```

```python
# convert labels into binary values

y[y == 2] = 0

y[y == 1] = 1
```

```python
# Check the data types of all columns in X
print(X.dtypes)
```

```python
# Apply one-hot encoding to categorical features in X
X = pd.get_dummies(X, drop_first=True)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, random_state = 0)
```

```python
# import XGBClassifier
from xgboost import XGBClassifier
# declare parameters
params = {
            'objective':'binary:logistic',
            'max_depth': 4,
            'alpha': 10,
            'learning_rate': 1.0,
            'n_estimators':100
        } 
            
# instantiate the classifier 
xgb_clf = XGBClassifier(**params)
# fit the classifier to the training data
xgb_clf.fit(X_train, y_train)
```

```python
# alternatively view the parameters of the xgb trained model
print(xgb_clf)
```

```python
# make predictions on test data
y_pred = xgb_clf.predict(X_test)
```

```python
# check accuracy score
from sklearn.metrics import accuracy_score

print('XGBoost model accuracy score: {0:0.4f}'. format(accuracy_score(y_test, y_pred)))
```
